```python
# Import Dependencies
import pandas as pd
```

```python
# Read in our unclean data
df = pd.read_csv('state_data.csv')
df.columns
```

```python
# Grab the only columns we want
labor_df = df[[
    'State',
    'Year',
    'Labor productivity (2012=100)',
    'Labor productivity (percent change)',
    'Real hourly labor compensation cost ($/hour)',
    'Real hourly labor compensation cost (percent change)',
    'Value of production (million $)'
]]

# Make a list of the only years that we want
year_list = [2015, 2016, 2017, 2018, 2019]

# And then grab only the years that we want from the data
clean_data = labor_df[labor_df['Year'].isin(year_list)]
clean_data.head(10)
```

```python
# Then we'll write our clean data to a csv
clean_data.to_csv('clean_data.csv', index= False)
```
